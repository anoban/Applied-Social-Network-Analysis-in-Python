```python
import networkx as nx
import numpy as np
```

```python
graph = nx.Graph()
graph.add_edges_from([("A", "B"), ("A", "C"), ("B", "D"), ("C", "D"), ("C", "E"), ("D", "E"), ("D", "G"), ("E", "G"), ("G", "F")])
nx.draw_networkx(graph, node_size = 800, node_color = "orange")
```

```python
nx.degree_centrality(graph).get("D")
```

```python
nx.closeness_centrality(graph).get("G")
```

```python
nx.betweenness_centrality(graph).get("G")
```

```python
nx.edge_betweenness_centrality(graph, normalized = False).get(("G", "F"))
```

```python
graph = nx.DiGraph()
graph.add_edges_from([("A", "B"), ("B", "A"), ("A", "C"), ("D", "C"), ("C", "D")])
nx.draw_networkx(graph, node_size = 800, node_color = "orange")
```

```python
for x in np.arange(0.4, 1, 0.05):
    print("Alpha = %.2f, PageRank(D) = %.8f" % (x, nx.pagerank(graph, alpha = x).get("D")))
```
