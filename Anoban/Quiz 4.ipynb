```python
def sort_dict_by_values(dictionary: dict) -> list:
    return sorted([item for item in dictionary.items()], reverse = True, key = lambda tup: tup[1])
```

```python
# Resource Allocation index of nodes C and D?

common_neighbors = set(nx.neighbors(graph, "C")).intersection(set(nx.neighbors(graph, "D")))
common_neighbors
```
