### ***Nodes or vertices -> Members in a network.***
### ***Edges -> Connections in between vertices.***

## *The edges (connections) can be of different types!*

- Non-directional edges (connections where the direction does not matter, symmetric relationships)         
e.g. A network of friends. If two peope A & B are friends, usually A is a friend of B and B is a friend of A. So, the edge here is non-directional. These can also be considered as bidirectional edges.

- Directed edges (connections where the direction of the edge is important, asymmetric relationships)
e.g A network of organisms in a foodweb. If two animals are invloved in a predator-prey relationship, the direction is critical to accurately ilustrate the relationship, if this scenario is represented by a non-directional graph

```python
import networkx as nx
import matplotlib.pyplot as plt
plt.rcParams["figure.figsize"] = (4, 4)
plt.rcParams["font.size"] = 10
```

## ***Undirected networks***

```python
# the Graph class is used for undirected graphs

graph = nx.Graph()
```

```python
type(graph)
```

```python
graph.add_edge("A", "B")
graph.add_edge("B", "C")
graph.add_edge("B", "D")
```

```python
nx.draw(graph, with_labels = True)
```

## ***Directed networks***

```python
# to represent directed graphs, we'll use directed graphs

dir_graph = nx.DiGraph()

# A -> B
dir_graph.add_edge("A", "B")

# A -> D
dir_graph.add_edge("A", "D")

# C -> B
dir_graph.add_edge("C", "B")

nx.draw(dir_graph, with_labels = True)
```

## ***Weighted networks***

```python
# weighted graphs
# in certain graphs, edges have different weights
# in a networks of socialites, the connections between two can be weighed based on the number of times they had lunched together.

# weighted graphs can be represented by the Graph class

weighted_graph = nx.Graph()

weighted_graph.add_edge("A", "B", weight = 4)
weighted_graph.add_edge("C", "B", weight = 2)
weighted_graph.add_edge("A", "D", weight = 1)
weighted_graph.add_edge("C", "A", weight = 6)

pos = nx.spring_layout(weighted_graph)
nx.draw(weighted_graph, with_labels = True)
nx.draw_networkx_edge_labels(weighted_graph, pos = pos, edge_labels = nx.get_edge_attributes(weighted_graph, "weight"));
```

## ***Signed networks***

```python
# networks can represent what types of relationships exist between individuals using signs
# e.g. friendship can be marked by a + sign while marking antagonism with a - sign.

signed_network = nx.Graph()

signed_network.add_edge("A", "B", sign = "+")
signed_network.add_edge("D", "B", sign = "-")
signed_network.add_edge("A", "C", sign = "-")
signed_network.add_edge("C", "B", sign = "+")

pos = nx.shell_layout(signed_network)
labs = nx.get_edge_attributes(signed_network, name = "sign")
nx.draw(signed_network, with_labels = True)
nx.draw_networkx_edge_labels(signed_network, pos = pos, edge_labels = labs);
```

```python
# in addition to weights & signs edges can aslo have other types of attributes.
# these are specified by the relation argument

graph = nx.Graph()

graph.add_edge("A", "B", relation = "Friends")
graph.add_edge("A", "B", relation = "Co-Workers")
graph.add_edge("C", "B", relation = "Neighbours")
graph.add_edge("A", "C", relation = "Family")
graph.add_edge("D", "C", relation = "Married")

labs = nx.get_edge_attributes(graph, name = "relation")
positions = nx.spring_layout(graph)
nx.draw(graph, with_labels = True)
nx.draw_networkx_edge_labels(graph, pos = positions, edge_labels = labs);
```
